```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Facebook = fb
fb = pd.read_excel('/content/FACEBOOK.xlsx')
fb
```

```python
# Instagram = ig
ig = pd.read_excel('/content/INSTAGRAM.xlsx')
ig
```

```python
# Linkedin = li
li = pd.read_excel('/content/LINKEDIN.xlsx')
li
```

```python
# X = x
x = pd.read_excel('/content/X.xlsx')
x
```

```python
# Make sure they are the same columns
print("FACEBOOK columns:", fb.columns.tolist())
print("INSTAGRAM columns:", ig.columns.tolist())
print("LINKEDIN columns:", li.columns.tolist())
print("X columns:", x.columns.tolist())
```

```python
# Check the number of rows in each file
print(len(fb), len(ig), len(li), len(x))
```

```python
# Data Appending
all = pd.concat([fb,ig,li,x], ignore_index=True)
all
```

```python
all.head()
```

```python
# Make sure to append
all.tail()
```

```python
# show total rows

print("Total Rows:", len(all))
print("Platforms:", all["Social_Media"].unique())
```

```python
# Display data information (Data Types)
all.info()
```

```python
# Statistical Summary
all.describe()
```

```python
# Make sure each platform has 52 rows
all["Social_Media"].value_counts()
```

```python
# Checking Missing Values
all.isnull().sum()
```

```python
# Calculating the average Engagement Rate for each platform
avg_engagement = all.groupby("Social_Media")["Engagement Rate"].mean()
print(avg_engagement)
# Knowing who is higher
highest_value = avg_engagement.max()
highest_platform = avg_engagement.idxmax()
print("The platform with the highest average engagement rate is:", highest_platform)
print("Average Engagement Rate:", highest_value)
```

```python
# Calculating the standard deviation for each platform
std_engagement = all.groupby("Social_Media")["Engagement Rate"].std()
print(std_engagement)
# Determining the most stable
most_consistent_platform = std_engagement.idxmin()
lowest_std = std_engagement.min()
print("The most consistent platform is:", most_consistent_platform)
print("Standard Deviation:", lowest_std)
```

```python
# Comparing all metrics across platforms
metrics_comparison = all.groupby("Social_Media").mean(numeric_only=True)
metrics_comparison
```

```python
#Q1: Compare average engagement rates across platforms
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(metrics_comparison.index, metrics_comparison['Engagement Rate'], color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel('Social Media Platform')
plt.ylabel('Average Engagement Rate')
plt.title('Average Engagement Rate by Social Media Platform')
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()
```

```python
# Q2: Analyze relationship between content visibility and interaction
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for platform in all['Social_Media'].unique():
    subset = all[all['Social_Media'] == platform]
    plt.scatter(subset['Impressions'], subset['Likes'], label=platform, alpha=0.7)

color=['skyblue', 'lightcoral', 'lightgreen', 'gold']

plt.xlabel('Social Media Platform')

plt.title('Impressions vs. Likes by Social Media Platform')

plt.xlabel('Total Impressions')

plt.ylabel('Total Likes')

plt.legend(title='Social Media')

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Q3: Track engagement patterns over time
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for platform in all['Social_Media'].unique():
    subset = all[all['Social_Media'] == platform]
    plt.plot(subset['Week'], subset['Engagement Rate'], label=platform)
color=['skyblue', 'lightcoral', 'lightgreen', 'gold']
plt.xlabel('Week')
plt.ylabel('Engagement Rate')
plt.title('Engagement Rate Over Time by Social Media Platform')
plt.legend(title='Social Media')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Q4: Compare audience growth variability
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for platform in all['Social_Media'].unique():
    subset = all[all['Social_Media'] == platform]
    plt.plot(subset['Week'], subset['Audience Growth Rate'], label=platform)
color=['skyblue', 'lightcoral', 'lightgreen', 'gold']
plt.xlabel('Week')
plt.ylabel('Audience Growth Rate')
plt.title('Audience Growth Rate Over Time by Social Media Platform')
plt.legend(title='Social Media')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Q5: Evaluate audience responsiveness
import matplotlib.pyplot as plt

df_avg_response_rate = all.groupby("Social_Media")["Response Rate"].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(df_avg_response_rate['Social_Media'], df_avg_response_rate['Response Rate'], color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel('Social Media Platform')
plt.ylabel('Average Response Rate')
plt.title('Average Response Rate by Social Media Platform')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
